```python
 # covid -19
import requests
from prettytable import PrettyTable

url = "https://covid-193.p.rapidapi.com/statistics"
headers = {
    "x-rapidapi-key": "25dc6f42a3mshe5b5b76c3c439f8p17c67bjsn8ddbdcd445fc",
    "x-rapidapi-host": "covid-193.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
result = response.json()

obj = PrettyTable()
obj.field_names = ["Country", "Cases", "Deaths", "Recovered", "Population"]

for item in result["response"]:
    country = item["country"]
    cases = item["cases"]["total"]
    deaths = item["deaths"]["total"]
    recovered = item["cases"]["recovered"]
    population = item["population"]

    obj.add_row([country, cases, deaths, recovered, population])

print(obj)
```
